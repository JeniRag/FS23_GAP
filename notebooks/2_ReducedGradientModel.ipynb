# Train and Test GAP model with defined number of forces

```python
massH = 1.6735575 * 1e-24 #gramm
angstrom_to_cm= 1e-8
```

Structure:
* First define SOAP Parametrs and create Kernel without gradient.
* Train on Energies only (find best lambda for energy)
* Predict on Energies only
* If Results are good, recompute Kernel with gradient (takes time)
* Train on Energies and Forces (find best labmda for forces)
* Predict on Energy and forces

## load data

```python
#training set
frame_folder="990_frames"
frames_train = ase_io.read(f"../frames_data/{frame_folder}/frames.xyz", index = ":")
force_idx=np.load(f"../frames_data/{frame_folder}/random_train_10indices.npy")

# test set
frames_test = ase_io.read(f"../frames_data/frames_test/frames.xyz", index = ":")
force_idx_test=np.load("../frames_data/frames_test/random_test_10indices.npy")
```

```python
## loading training energies and forces
energies_train = np.array([x.get_total_energy() for x in frames_train])
forces_train = np.array([x.get_forces() for x in frames_train])
forces_red_train = reduce_forces(forces_train, force_idx) #instead of using all forces use only reduced number of forces per frame

for fr in frames_train:
    fr.wrap(eps=1e-12)  

## loading testing energies and forces
energies_test = np.array([x.get_total_energy() for x in frames_test])
forces_test = np.array([x.get_forces() for x in frames_test])
forces_red_test = reduce_forces(forces_test, force_idx_test)

for frt in frames_test:
    frt.wrap(eps=1e-12)
    
#additional frame data for later plottings
volumes_test = [frames.get_volume() for frames in frames_test]
number_of_atoms_test = [len(frames) for frames in frames_test]
volumes_test= np.array(volumes_test) * angstrom_to_cm**3 #in cm3
density_test = massH *np.array(number_of_atoms_test) / volumes_test # in g/cm3
```

```python
natoms = len(frames_train[0])
mean_nrg_per_atom = np.mean(energies_train/natoms)
mean_nrg = natoms* mean_nrg_per_atom

#train_energies = energies_train - mean_nrg
```

```python
global_species = set()
for frame in frames_train:
    global_species.update(frame.get_atomic_numbers())
global_species = np.array(list(global_species))

# Atomic energy baseline
atom_energy_baseline = np.mean(energies_train)/(frames_train[0].get_global_number_of_atoms())

energy_baseline = {int(species): atom_energy_baseline for species in global_species}
```

Note that I don't subtract the mean energy from the training energy, since it is done in train_gap_model(), if the self_contribution term is provided.

# Build Kernel without gradient

## build kernels

Define Soap parameters

```python
HYPERS={'soap_type': 'PowerSpectrum',
     'interaction_cutoff': 3.0,
     'max_radial': 6,
     'max_angular': 4,
     'gaussian_sigma_constant': 0.2,
     'gaussian_sigma_type': 'Constant',
     'cutoff_smooth_width': 0.5,
     'radial_basis': 'GTO',
     'normalize': True,
     'cutoff_function_type': 'RadialScaling',
     'cutoff_function_parameters': {'rate': 4.0,
      'scale': 1.0,
      'exponent': 4,
      'interaction_cutoff': 3.0,
      'cutoff_smooth_width': 0.5}}
```

```python
## create the SOAP object
soap = SphericalInvariants(**HYPERS)
```

Convert to soap representation

```python
## generate the managers
train_managers = soap.transform(frames_train) 
```

Select Sparse Points

```python
## define the Sparse points
## can be slow depending on the number of sparse points
n_sparse = {1: 1000} 
zeta = 2

sparse = FPSFilter(soap, n_sparse, act_on='sample per species')
X_sparse = sparse.select_and_filter(train_managers)

#save
dump_obj(f"../intermediate_data/{folder}/X_sparse.json", X_sparse)
```

```python
kernel = Kernel(soap, 
                name='GAP', 
                zeta=zeta, 
                target_type='Structure', 
                kernel_type='Sparse')
#save
dump_obj(f"../intermediate_data/{folder}/kernel.json", kernel)
```

```python
kNM = kernel(train_managers, X_sparse) 
print(f"KNM shape: {kNM.shape}")
```

## find optimal regularizer

```python
errors_tmp_file=f"../intermediate_data/{folder}/tmp/errors_reg1_tmp.npy" #check_point file
Restart=False #in case kernel died during computation, start from check point file
```

```python
cv = 2
kfold = KFold(n_splits=cv, shuffle=True, random_state=10)
```

```python
reg_arr = np.logspace(-2, 5, 16) 
reg_arr = np.unique( np.append(reg_arr, 1.0))
```

```python
#if restarted, load the intermediate results
if (os.path.exists(errors_tmp_file) == True) & Restart==True:
    errors = list(np.load(errors_tmp_file))
    start_index = len(errors)
 
# if it doesn't exist, create empty list
else:
    errors = []
    start_index = 0

## optimization loop for the energies
for regularization in tqdm(reg_arr[start_index:]):
   
    temp_err = 0.
    for train, test in kfold.split(range(len(energies_train))): 
        fr_train = [frames_train[j] for j in train]
        
        model = train_gap_model(kernel, fr_train, kNM[train], X_sparse, energies_train[train], energy_baseline, solver = "RKHS-QR",
                     lambdas=[regularization], jitter=0)
        
        target_pred = kNM @ model.weights 
        target_pred += mean_nrg #model.self_contributions[1] *128 #+ energy_baseline[1]
        temp_err += get_rmse(target_pred[test], 
                             energies_train[test],
                             perc=True)
    errors.append(temp_err/cv)

    #store intermediate data
    np.save(errors_tmp_file, errors)
    
errors = np.asarray(errors)

#store array and errors
error_and_range = np.zeros(shape=(2, len(reg_arr)) )
error_and_range[0,:]= reg_arr
error_and_range[1,:] = errors
np.save(f"../intermediate_data/{folder}/error_and_range_reg1.npy", error_and_range)

#regularization value with lowest RMSE value
regularization1 = reg_arr[errors.argmin()]
print('energy lambda = {}'.format(regularization1))
np.save(f"../intermediate_data/{folder}/reg1.npy", regularization1)
```

```python
#error_and_range = np.load(f"../intermediate_data/{folder}/error_and_range_reg1.npy")
plt.loglog(error_and_range[0,:], error_and_range[1,:], "o-")
plt.ylabel("%RMSE")
plt.xlabel("regularization")
plt.title(f"minimum RMSE: {errors.min()} \n reg1 = {regularization1}" )
plt.savefig(f"../intermediate_data/{folder}/images/reg1.png", dpi = 300)
```

## build kernel for train set

```python
RESTART={"status": False, #if you want to restart computation from checkpoint file
        "restart_index": 720}

save_N=200 #save after every save_N steps
```

```python
#if you restart from intermediate results
if RESTART["status"]:
    #load intermediate Kernels
    restart_index = RESTART["restart_index"]
    kNM = np.load(f"../intermediate_data/{folder}/kenrel_chk/kNM_train_chk_{restart_index}.npy")
    kNM_grad = np.load(f"../intermediate_data/{folder}/kernel_chk/kNM_grad_train_chk_{restart_index}.npy")
    N=kNM.shape[0]
    
    #convert np.ndarray into list
    kNM=kNM.tolist()
    kNM_grad=kNM_grad.tolist()

else:   
    kNM = []
    kNM_grad =[]
    N=0

for i in tnrange(N, len(frames_train)):
    k=compute_KNM([frames_train[i]], X_Sparse, kernel_grad, soap_grad)
    kNM.append(k[0])
    for j in range(force_idx.shape[1]):
        kNM_grad.append(k[1:][force_idx[i,j]*3:force_idx[i,j]*3+3])
    
    if i % save_N ==0:
        np.save(f"../intermediate_data/{folder}/kernel_chk/kNM_train_chk_{i}.npy", kNM)
        np.save(f"../intermediate_data/{folder}/kernel_chk/kNM_grad_train_chk_{i}.npy", kNM_grad)
      

np.save(f"../intermediate_data/{folder}/kernel_chk/kNM_train_chk_{i}.npy", kNM)
np.save(f"../intermediate_data/{folder}/kernel_chk/kNM_grad_train_chk_{i}.npy", kNM_grad)


kNM_g=np.concatenate(np.array(kNM_grad), axis=0)
del kNM_grad

#stack energy and forces kernels together
kNM_tot=np.concatenate([kNM, kNM_g], axis=0)

np.save(f"../intermediate_data/{folder}/kNM_train.npy", kNM_tot)
assert kNM_tot.shape == (len(frames_train) + len(frames_train)*force_idx.shape[1]*3, len(X_Sparse))
print(np.shape(kNM_tot))
```

```python
#to free memory
del kNM_g
del kNM_tot
del RESTART
```

## find optimal regularizer

```python
Restart=False
errors_tmp_file=f"../intermediate_data/{folder}/tmp/errors_reg2_tmp.npy"
rate_energy = 0.5
```

```python
cv = 2
kfold = KFold(n_splits=cv, shuffle=True, random_state=20)
```

```python
reg_arr = np.logspace(-1, 5, 16) #18 min (-10,10, 31)
reg_arr = np.unique( np.append(reg_arr, 1.0))
```

```python
if (os.path.exists(errors_tmp_file) == True) & Restart==True:
        errors = list(np.load(errors_tmp_file))
        start_index = len(errors)
        
else:
    errors = []
    start_index = 0
    
    
for regularization2 in tqdm(reg_arr[start_index:]):
  
    temp_err = 0.
    for train, test in kfold.split(range(len(frames_train))):
    
        fr_train = [frames_train[i] for i in train]
        kNM_train_train = extract_KNM(frames_train, kNM_train, train, Nforces =force_idx.shape[1])
        
        model = train_gap_model(kernel_grad, fr_train, kNM_train_train, X_Sparse, energies_train[train], energy_baseline, solver = "RKHS-QR",
                    grad_train = -forces_red_train[train], 
                    lambdas=[regularization1, regularization2], jitter=0)
        
        target_pred = kNM_train[:len(frames_train)] @ model.weights
        target_pred += mean_nrg
        
        target_grad_pred = predict_forces(model, frames_train, kNM_train, force_idx)
        
        tmp = rate_energy * get_rmse(target_pred[test], 
                            energies_train[test],
                            perc=True)
        
        tmp += (1-rate_energy) * get_rmse(target_grad_pred[test], 
                            forces_red_train[test],
                            perc=True)
        temp_err += tmp
        
    errors.append(temp_err/cv)

    np.save(errors_tmp_file, errors)

errors = np.asarray(errors)

error_and_range = np.zeros(shape=(2, len(reg_arr)))
error_and_range[0,:] = reg_arr
error_and_range[1, :] =errors
np.save(f"../intermediate_data/{folder}/error_and_range_reg2.npy", error_and_range)

reg2 = reg_arr[errors.argmin()]
print('force lambda reg2 = {}'.format(reg2))
np.save(f"../intermediate_data/{folder}/reg2_{rate_energy}.npy", reg2)
```

```python
plt.loglog(error_and_range[0,:], error_and_range[1,:], "o-")
plt.xlabel("regularization")
plt.ylabel("%RMSE")
plt.title(f"minimum RMSE: {errors.min()} \n reg2 = {reg2}")
plt.savefig(f"../intermediate_data/{folder}/images/reg2_{rate_energy}.png", dpi =300)
```
